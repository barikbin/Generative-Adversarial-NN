```python
from tensorflow.python.ops.parallel_for.control_flow_ops import for_loop

import os
import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# This is the normal Loading
# Print out a graph with 1 Add node

x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
z = tf.add(x, y)

with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        writer = tf.summary.FileWriter('graphs/lazy_loading', sess.graph)
        for _ in range(10):
                sess.run(tf.add(x,y))
        print(tf.get_default_graph().as_graph_def())
        writer.close()
                
```

```python

## Lazy Loading
## print out a graph with 10 Adde nodes

x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')

with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        writer = tf.summary.FileWriter('graphs/lazy_loading', sess.graph)
        for _ in range(10):
                 sess.run(tf.add(x,y))
        print(tf.get_default_graph().as_graph_def())
        writer.close()   


# Key take ways is that nodes and graph have a totally different meaning!
```
