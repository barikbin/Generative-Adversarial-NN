```python

import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# this is a fee_dict with placeholder
# This is a placeholder for a vector of 3 elements, type tf.floar32

a = tf.placeholder(tf.float32, shape=[3])
b = tf.constant([5,5,5], tf.float32)
```

```python
c = a + b
writer = tf.summary.FileWriter('graph/placeholders', tf.get_default_graph())
with tf.Session() as sess:
   
  # compute the value of c given the value of a is [1, 2, 3] equals to [6. 7. 8.]
  # To get the result we need to run this syntax

  print(sess.run(c,{a: [1, 2, 3]}))

# Pay attention that we need to close the writer!
writer.close()

a = tf.add(2,5)
b = tf.multiply(a, 3)

with tf.Session() as sess:

  # we run the calculation two times here
  # 10 *
   print(sess.run(b))

# This is tricky syntax!  we multiply a three times.. a * 3b
   print(sess.run(b, feed_dict={a: 15}))

# need to redo the operand in tailwind css!
```

```python
# Different calculation
# Here we create simple log file writer

import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

a = tf.constant(2, name='a')
b = tf.constant(3, name='b')
x = tf.add(a, b, name='add')
writer = tf.summary.FileWriter('./graphs/simple', tf.get_default_graph())

# This is an import syntax
with tf.compat.v1.Session() as sess:
    print(sess.run(x))
writer.close()
```

```python

# Example number 2
a = tf.constant([2,2], name='a')

# Notice the shape of tensor
b = tf.constant([[0,1], [2,3]], name='b')

with tf.compat.v1.Session() as sess:
    print("First example")
    print(sess.run(tf.div(b,a)))
    print("Second example")
    print(sess.run(tf.math.divide(b,a)))
    print("Third example")
    print(sess.run(tf.math.truediv(b,a)))
    print("Fourth example")
    print(sess.run(tf.floordiv(b,a)))
    print("Fifth example")
    print(sess.run(tf.truncatediv(b, a)))
    print("sixth example")
    print(sess.run(tf.floor_div(b,a)))
```
